```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])  #Y values. #So by default indexes become 0,1,2,3
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.title('MyPlot')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16]) #x values corresponding to y
plt.ylabel("Numbers")
plt.xlabel("Squared")
plt.title('MyPlot')
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro') #r is the color red and o shape dots.
plt.ylabel("Numbers")
plt.xlabel("Squared")
plt.title('MyPlot')
plt.grid()
plt.show()
```

```python
import numpy as np
t = np.arange(1.,5.,0.2)
plt.plot(t, t**2, 'b--', label='^2')
plt.plot(t, t**2.2, 'rs', label='^2')
plt.plot(t, t**2.5, 'g^', label='^2')
plt.grid()
plt.legend()    #Shows representation
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],linewidth=5.0)  #Line width
plt.show()
```

**Set Properties:**

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]
lines = plt.plot(x1,y1,x2,y2)

#Use keyword args
plt.setp(lines[0], color='r', linewidth=2.0)

# or MATLAB style string value pairs

plt.setp(lines[1],'color','g','linewidth',5.0)
plt.grid()
```

**Working with multiple figures and axis:**

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)  #Subplot() specifies number of rows n cols. 211 means 2 rows 1 col 1st subplot
#Subplot breaks whole figure in to parts and we specify (rows,cols,part no.)
plt.grid()
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'b-')
plt.show()
```

```python
plt.figure(1) #1st fig
plt.subplot(211)  #1st subploit in fig 1
plt.plot([1,2,3]) 
plt.subplot(212)  #2nd subploit in fig 1
plt.plot([4,5,6])

plt.figure(2)  #2nd fig
plt.plot([4,5,6])  #creates subplot 111 by default

plt.figure(1)  #fig 1 current; subplot(212) still current
plt.subplot(211)  #make subplot(211) in current fig
plt.title('Easy as 1,2,3')  #subplot 211 title
plt.show()
```
